# 2. Merge both data frames on movieid

```python
df_combined = pd.merge(ratings, movies, on='movieId')
df_combined
```

```python
df_combined.info()
```

```python
ratings_matrix = df_combined.pivot_table(index='userId', columns='title',values='rating')
ratings_matrix
```

# 4. Perform User-based Collaborative Filtering

### 4.1 Fill the missing values with the average ratings

Use User-base collaborative filtering

```python
user_utility_matrix = ratings_matrix.copy()
user_utility_matrix = user_utility_matrix.apply(lambda row: row.fillna(row.mean()), axis=1)
user_utility_matrix
```

# 4.2 Choose the correlation of all users with User 1 only

We need to transpose the matrix in order to capture the Users information (the rows)

```python
user_corr_matrix =  user_utility_matrix.T.corr()
user_corr_matrix
```

```python
corr_user_1 = user_corr_matrix.iloc[0]
```

```python
corr_user_1.sort_values(ascending=False, inplace=True)
corr_user_1
```

The closest users are: 301 and 597 and 414

```python
top50_users = corr_user_1[1:51]
top50_users
```

# 4.6 Predict the rating that User 1 might give for the movie with movieId 32 based on the top 50 user correlation matrix

{Hint: Predicted rating = sum of [(weights) * (ratings)] / sum of (weights )}.
Here, weights are the correlation of the corresponding user with the first user.). 
That is, the predicted rating is calculated as the weighted average of k similar users


```python
# Calculate Predicted rating = sum of [(weights) * (ratings)] / sum of (weights )
```

```python
sum_similarity = 0
weighted_ratings = 0

for user in top50_users.index:
    sum_user = df_combined[(df_combined['userId'] == user) & (df_combined['movieId'] == 32)]['rating'].sum()
    weighted_ratings += top50_users.loc[user] * sum_user
    sum_similarity += top50_users.loc[user]
    
print(weighted_ratings / sum_similarity)
```

Downside of this approach: Everytime a user rates a movie, you have to go back and do all the calculations.


This can become very computational intensive

Item-based Collaborative filtering is a more recent approach to overcome this limitations

# 5. Perform Item-based Collaborative Filtering

### 5.1 Fill the column-wise NaN's in the User-Item Matrix with the corresponding movie's mean ratings, and find Pearson correlation between movies

Tell which items are similar (instead of users)

```python
item_utility_matrix = ratings_matrix.copy()
item_utility_matrix = item_utility_matrix.apply(lambda col: col.fillna(col.mean()), axis=0)
item_utility_matrix
```

• Choose the correlation of all movies with the movie Jurassic Park (1993) only.
• Sort the  Jurassic Park movie correlation in descending order
• Drop the NaN values generated in the correlation matrix
• Find the most 10 movies similar to the movie Jurassic Park (1993)

```python
item_corr_matrix = item_utility_matrix.corr()
item_corr_matrix
```

```python
item_corr_matrix.dropna(inplace=True)
item_corr_matrix
```

### 5.2 Choose the correlation of all movies with the movie Jurassic Park (1993) only.

```python
similar_movies = item_corr_matrix[['Jurassic Park (1993)']].sort_values(ascending=False, by='Jurassic Park (1993)')
```

```python
top10_items = similar_movies[1:11]
print('top10 similar movies: ', top10_items)
```

# 6. Perform KNNBasic, SVD, NMF Model-based Collaborative Filtering

    Initialize KNNBasic with similarity configuration as Mean Squared Distance Similarity 
(msd), 20 neighbors and cross-validate 5 folds against measure RMSE.
    (Hint: cross_validate(algo=algo, data=data, measures=['RMSE'], cv=5, verbose=True)) 

    Initialize Singular Value Decomposition (SVD) and cross-validate 5 folds against measure RMSE.

    Initialize Non-Negative Matrix Factorization (NMF)and cross-validate 5 folds against measure RMSE.

    Print best score and best params from Cross Validate on all the models built.


```python
from surprise import Reader, Dataset, KNNBasic, SVD, NMF
from surprise.model_selection import GridSearchCV, cross_validate
```

```python
# Range of the rating values
reader = Reader(rating_scale=(0.5, 5.0))

data = Dataset.load_from_df(ratings[['userId','movieId','rating']], reader=reader)
```

```python
data
```

```python
sim_options = {'name': 'msd'}

algo = KNNBasic(k=20, sim_options=sim_options)
# Do 5 traingin rounds (folds)
cross_validate(algo=algo, data=data, measures=['RMSE'], cv=5, verbose=True)
```

### Initialize Singular Value Decomposition (SVD) and cross-validate 5 folds against measure RMSE.

```python
algo = SVD()
cross_validate(algo=algo,data=data, measures=['RMSE'],cv=5,verbose=True)
```

```python
param_grid = {'n_factors': [50, 75], 'lr_all':[0.5, 0.05], 'reg_all': [0.06, 0.04]}

#Use grid search for parameter tuning
#SVD model was imported from surprise library
gs = GridSearchCV(algo_class=SVD, measures=['RMSE'], param_grid=param_grid)
gs.fit(data)
```

```python
print("Best score", gs.best_score['rmse'])
print("Best parameters", gs.best_params['rmse'])
```

### Initialize Non-Negative Matrix Factorization (NMF)and cross-validate 5 folds against measure RMSE.

```python
algo = NMF()
cross_validate(data=data, algo=algo, measures=['RMSE'], cv=5, verbose=True)
```

# Make model predictions to recommend movies to the users

```python
from surprise.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(data, test_size=0.25)

model = algo.fit(train_data)

predictions = model.test(test_data)

predictions
# get the RMSE to test predictions
```
